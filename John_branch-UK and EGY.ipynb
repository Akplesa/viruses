```python
import csv 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
%matplotlib inline

#read CSV files that contain pandemic data 
COVID_data = pd.read_csv("data_sets/COVID-19.csv") 
COVID_data


#5 countries of focus
#AP= US, Australia    ZL= China, Russia  JB= GBR, EGY

#Clean COVID-19 DataFrame and extract data based on location, date, total_cases, and total_deaths
#Create new COVID-19 DataFrame
#Multiple Line graph for date and total_cases for each country
#Line graph for date and total deaths for each country
```

```python
#Clean COVID-19 DataFrame and extract data based on location, date, total_cases, and total_deaths

clean_data_c = COVID_data[["iso_code", "location", "date", "total_cases", "total_deaths", "new_deaths", "new_cases"]]
clean_data_c
#Multiple Line graph for date and total_cases for each country
GBR_c = clean_data_c[(clean_data_c["iso_code"] == 'GBR')]


x_axis_GBR_c = GBR_c["date"]
y_axis_GBR_c = GBR_c["total_cases"]
GBR_c_cases_total = plt.plot(x_axis_GBR_c, y_axis_GBR_c, color= "blue", linewidth=1, label="GBR")
plt.xticks(rotation = "90")
plt.tight_layout()
plt.xticks(np.arange(0, len(x_axis_GBR_c)+1, 9))
plt.legend(["Great Britain", "Egypt"])
plt.title("Total COVID Cases in Great Britain and Egypt")
plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.tight_layout()
plt.show()




```

```python
EGY_c = clean_data_c[(clean_data_c["iso_code"] == 'EGY')]
x_axis_EGY_c = EGY_c["date"]
y_axis_EGY_c = EGY_c["total_cases"]
EGY_cases_total = plt.plot(x_axis_EGY_c, y_axis_EGY_c, color= "green", linewidth=1, label="EGY")
plt.xticks(rotation = "90")
plt.tight_layout()
plt.xticks(np.arange(0, len(x_axis_EGY_c)+1, 9))
plt.legend(["Great Britain", "Egypt"])
plt.title("Total COVID Cases in Great Britain and Egypt")
plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.tight_layout()
plt.show()
```

```python
##COVID-19 Do more cases= more deaths?
# plot scatter plot and linear regression for each country Cases x, Deaths y

cases_G_c = np.array(GBR_c["total_cases"])
deaths_G_c = np.array(GBR_c["total_deaths"])
f = plt.figure(figsize=(5,5))
ax = f.add_subplot(111)
plt.plot(cases_G_c, deaths_G_c, '.')

m,b = np.polyfit(cases_G_c, deaths_G_c, 1)
plt.plot(cases_G_c,m*cases_G_c+b)
#GBR_cases = plt.scatter(x_axis_GBR, y_axis_GBR, color= "red", facecolors="red", edgecolors="black", label="GBR")


plt.tight_layout()
#plt.xticks(np.arange(0, len(cases_G_c)+1, 1000))
#plt.legend(["Great Britain", "Egypt"])
plt.title("COVID Pandemic in Great Britain: Total Cases vs Total Deaths")
plt.xlabel("Total Cases")
plt.ylabel("Total Deaths")
plt.tight_layout()
plt.show()
```

```python
cases_E_c = np.array(EGY_c["total_cases"])
deaths_E_c = np.array(EGY_c["total_deaths"])
f = plt.figure(figsize=(5,5))
ax = f.add_subplot(111)
plt.plot(cases_E_c, deaths_E_c, '.')
m,b = np.polyfit(cases_E_c, deaths_E_c, 1)
plt.plot(cases_E_c,m*cases_E_c+b)
#GBR_cases = plt.scatter(x_axis_GBR, y_axis_GBR, color= "red", facecolors="red", edgecolors="black", label="GBR")


plt.tight_layout()
#plt.xticks(np.arange(len(x_axis_EGY_c)+1)
#plt.legend(["Great Britain", "Egypt"])
plt.title("COVID Pandmemic Egypt: Total Cases vs Total Deaths")
plt.xlabel("Total Cases")
plt.ylabel("Total Deaths")
plt.tight_layout()
plt.show()
```
