```python
#read CSV files that contain pandemic data 
COVID_data = pd.read_csv("data_sets/COVID-19.csv") 
COVID_data.head()
```

```python
#LOCATING COVID DATA
Covid_China = COVID_data.loc[COVID_data['iso_code'] == 'CHN']
Covid_Russia = COVID_data.loc[COVID_data['iso_code']=='RUS']
```

```python
#display the dataframe
Covid_China.head()
```

```python
Covid_China.columns
```

```python
Covid_China = Covid_China[['date','total_cases','new_cases','total_deaths','new_deaths','total_tests','new_tests']]
Covid_China
```

```python
##COVID-19 Do more cases= more deaths?
# plot scatter plot and linear regression for each country Cases x, Deaths y
#plot the scatter plot for China
case_C = np.array(Covid_China['new_cases'])
death_C = np.array(Covid_China['new_deaths'])
f = plt.figure(figsize=(5,5))
ax = f.add_subplot(111)
plt.plot(case_C,death_C,'.')
m,b = np.polyfit(case_C,death_C,1)
plt.plot(case_C,m*case_C+b)
plt.xlabel('new cases')
plt.ylabel('new deaths')
plt.tight_layout()
```

```python
import scipy.stats as ss
r,p = ss.pearsonr(case_C,death_C)
print(f'The correlation is {r} and the P-value is {p}')
```

```python
#plot the scatter plot for Russia
case_R = np.array(Covid_Russia['new_cases'])
death_R = np.array(Covid_Russia['new_deaths'])
f = plt.figure(figsize=(5,5))
ax = f.add_subplot(111)
plt.plot(case_R,death_R,'.')
m,b = np.polyfit(case_C,death_C,1)
plt.plot(case_C,m*case_C+b)
plt.xlabel('new cases')
plt.ylabel('new deaths')
plt.tight_layout()
```

```python
import scipy.stats as ss
r,p = ss.pearsonr(case_R,death_R)
print(f'The correlation is {r} and the P-value is {p}')
```
