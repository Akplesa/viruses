```python
##COVID-19 Do more cases= more deaths?
# plot scatter plot and linear regression for each country Cases x, Deaths y
#plot the scatter plot for China
from scipy.stats import linregress
x_values = Covid_China['total_cases']
y_values = Covid_China['total_deaths']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('COVID-19 Cases')
plt.ylabel('COVID-19 Deaths')
plt.title("COVID-19 Cases v.s Deaths in the China")
print(f"The r-squared is: {rvalue**2}")
plt.legend(['R^2= 0.98'])
plt.show()
```

```python
import scipy.stats as ss
r,p = ss.pearsonr(x_values,y_values)
print(f'The correlation is {r} and the P-value is {p}')
```
