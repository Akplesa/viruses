```python
#US CASES VS DEATHS

# Add the linear regression equation and line to plot

x_values = US['total_cases']
y_values = US['total_deaths']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('COVID-19 Cases')
plt.ylabel('COVID-19 Deaths')
plt.title("COVID-19 Cases v.s Deaths in the U.S")
print(f"The r-squared is: {rvalue**2}")
plt.legend(['R^2= 0.98'])
plt.show()

# 98% of the variation in one variable may be explained by the other. This is showing that there is a strong positive correlation 
#between COVID cases and Deaths in the US. 
```

```python


##Hypothesis: Covid-19 will have a higher transmition rate than H1N1. 

# Independent T-test 
# Generate some fake data to test with
def gendata(loc1=0, loc2=0):
    population1 = stats.norm.rvs(loc=loc1, size=39493, random_state=0)
    population2 = stats.norm.rvs(loc=loc2, size=37419, random_state=42)

    # Scatter Plot of Data
    plt.subplot(2, 1, 1)
    plt.scatter(range(len(population1)), population1, label="COVID-19 average cases")
    plt.scatter(range(len(population2)), population2, label="H1N1 average cases")
    plt.legend()

    # Histogram Plot of Data
    plt.subplot(2, 1, 2)
    plt.hist(population1, 10, density=True, alpha=0.7, label="COVID-19 average cases")
    plt.hist(population2, 10, density=True, alpha=0.7, label="H1N1 average cases")
    plt.axvline(population1.mean(), color='k', linestyle='dashed', linewidth=1)
    plt.axvline(population2.mean(), color='k', linestyle='dashed', linewidth=1)
    plt.legend()  
    
    return population1, population2
```

```python

# Generate data with a bigger difference in means
population1, population2 = gendata(loc1=0, loc2=1)
# Note: Setting equal_var=False performs Welch's t-test which does 
# not assume equal population variance
stats.ttest_ind(population1, population2, equal_var=False)
```
